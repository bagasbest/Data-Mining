# Pie Chart

```python
#Sampel kecamatan yang di ambil secara acak
jagakarsa = df_drop.loc[df_drop['KECAMATAN'] == 'Jagakarsa'].count()[0]
menteng = df_drop.loc[df_drop['KECAMATAN'] == 'Menteng'].count()[0]
cilandak = df_drop.loc[df_drop['KECAMATAN'] == 'Cilandak'].count()[0]
cilincing = df_drop.loc[df_drop['KECAMATAN'] == 'Cilincing'].count()[0]
penjaringan = df_drop.loc[df_drop['KECAMATAN'] == 'Penjaringan'].count()[0]

labels = ['Jagakarsa', 'Menteng', 'Cilandak', 'Cilincing', 'Penjaringan']
colors = ['red', 'green', 'yellow', 'magenta', 'purple']

plt.pie([jagakarsa, menteng, cilandak, cilincing, penjaringan], labels=labels, colors=colors, autopct='%.2f %%')
plt.show()

print("Keterangan : ")
print("Penjaringan ada ", penjaringan, "Sekolah")
print("Cilincing ada ", cilincing, "Sekolah")
print("Jagakarsa ada ", jagakarsa, "Sekolah")
print("Cilandak ada ", cilandak, "Sekolah")
print("Menteng ada ", menteng, "Sekolah")
```
